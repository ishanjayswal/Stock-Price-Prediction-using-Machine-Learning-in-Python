```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from tqdm import tqdm

def add_technical_indicators(data):
    # MACD
    exp1 = data['Close'].ewm(span=12, adjust=False).mean()
    exp2 = data['Close'].ewm(span=26, adjust=False).mean()
    data['MACD'] = exp1 - exp2
    data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()

    # Bollinger Bands
    data['MA20'] = data['Close'].rolling(window=20).mean()
    data['SD20'] = data['Close'].rolling(window=20).std()
    data['Upper_BB'] = data['MA20'] + (data['SD20'] * 2)
    data['Lower_BB'] = data['MA20'] - (data['SD20'] * 2)
    
    # Stochastic Oscillator
    low_14 = data['Low'].rolling(window=14).min()
    high_14 = data['High'].rolling(window=14).max()
    data['%K'] = (data['Close'] - low_14) * 100 / (high_14 - low_14)
    data['%D'] = data['%K'].rolling(window=3).mean()

    # ATR
    data['H-L'] = data['High'] - data['Low']
    data['H-PC'] = abs(data['High'] - data['Close'].shift(1))
    data['L-PC'] = abs(data['Low'] - data['Close'].shift(1))
    data['TR'] = data[['H-L', 'H-PC', 'L-PC']].max(axis=1)
    data['ATR'] = data['TR'].rolling(window=14).mean()

    return data

# Fetch historical data
stock = 'ICICIBANK.NS'
data = yf.download(stock, start='2002-07-01', end='2024-09-10')

# Add technical indicators
data = add_technical_indicators(data)

# Existing features
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()

delta = data['Close'].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data['RSI'] = 100 - (100 / (1 + rs))

data['Close_lag1'] = data['Close'].shift(1)
data['Close_lag2'] = data['Close'].shift(2)
data['Close_lag3'] = data['Close'].shift(3)

data['Day'] = data.index.day
data['Month'] = data.index.month
data['Day_of_week'] = data.index.dayofweek

data['Target'] = data['Close'].shift(-1)

data.dropna(inplace=True)

# Features and target
features = ['Close', 'SMA_50', 'SMA_200', 'RSI', 'Close_lag1', 'Close_lag2', 'Close_lag3', 'Volume', 'Day', 'Month', 'Day_of_week',
            'MACD', 'Signal_Line', 'Upper_BB', 'Lower_BB', '%K', '%D', 'ATR']
X = data[features]
y = data['Target']

# Scale the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Initialize TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)

# Initialize the Random Forest Regressor
model = RandomForestRegressor(n_estimators=500, max_depth=15, min_samples_split=10, min_samples_leaf=5, random_state=42)

# Perform cross-validation
mse_scores = []
for train_index, test_index in tqdm(tscv.split(X_scaled), total=tscv.n_splits, desc="Cross-validation"):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Print cross-validation results
print(f"Cross-validation MSE scores: {mse_scores}")
print(f"Average MSE: {np.mean(mse_scores)}")
print(f"Standard deviation of MSE: {np.std(mse_scores)}")

# Train the model on the entire dataset
model.fit(X_scaled, y)

# Make predictions
y_pred = model.predict(X_scaled)

# Plot the predicted vs actual values
plt.figure(figsize=(12,6))
plt.plot(data.index, y, label='Actual Price', color='blue')
plt.plot(data.index, y_pred, label='Predicted Price', color='orange')
plt.title(f"{stock} Stock Price Prediction with Enhanced Features")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='best')
plt.show()

# Feature importance
feature_importance = pd.DataFrame({'feature': features, 'importance': model.feature_importances_})
feature_importance = feature_importance.sort_values('importance', ascending=False)
print("\
Feature Importance:")
print(feature_importance)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from tqdm import tqdm

def add_technical_indicators(data):
    # MACD
    exp1 = data['Close'].ewm(span=12, adjust=False).mean()
    exp2 = data['Close'].ewm(span=26, adjust=False).mean()
    data['MACD'] = exp1 - exp2
    data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()

    # Bollinger Bands
    data['MA20'] = data['Close'].rolling(window=20).mean()
    data['SD20'] = data['Close'].rolling(window=20).std()
    data['Upper_BB'] = data['MA20'] + (data['SD20'] * 2)
    data['Lower_BB'] = data['MA20'] - (data['SD20'] * 2)
    
    # Stochastic Oscillator
    low_14 = data['Low'].rolling(window=14).min()
    high_14 = data['High'].rolling(window=14).max()
    data['%K'] = (data['Close'] - low_14) * 100 / (high_14 - low_14)
    data['%D'] = data['%K'].rolling(window=3).mean()

    # ATR
    data['H-L'] = data['High'] - data['Low']
    data['H-PC'] = abs(data['High'] - data['Close'].shift(1))
    data['L-PC'] = abs(data['Low'] - data['Close'].shift(1))
    data['TR'] = data[['H-L', 'H-PC', 'L-PC']].max(axis=1)
    data['ATR'] = data['TR'].rolling(window=14).mean()

    return data

def generate_signals(data, period):
    data['Future_Close'] = data['Close'].shift(-period)
    data['Signal'] = np.where(data['Future_Close'] > data['Close'], 1, 0)  # Buy (1) if future close is higher
    data['Label'] = np.where(data['Signal'] == 1, 'Buy', 'Sell')  # Simplified example
    return data

# Fetch historical data
stock = 'ICICIBANK.NS'
data = yf.download(stock, start='2002-07-01', end='2024-09-10')

# Add technical indicators
data = add_technical_indicators(data)

# Add additional features
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()

delta = data['Close'].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data['RSI'] = 100 - (100 / (1 + rs))

data['Close_lag1'] = data['Close'].shift(1)
data['Close_lag2'] = data['Close'].shift(2)
data['Close_lag3'] = data['Close'].shift(3)

data['Day'] = data.index.day
data['Month'] = data.index.month
data['Day_of_week'] = data.index.dayofweek

# Drop rows with NaN values
data.dropna(inplace=True)

# Generate trading signals
period = 90  # Example: 90 days (3 months)
data = generate_signals(data, period)

# Features and target
features = ['Close', 'SMA_50', 'SMA_200', 'RSI', 'Close_lag1', 'Close_lag2', 'Close_lag3', 'Volume', 'Day', 'Month', 'Day_of_week',
            'MACD', 'Signal_Line', 'Upper_BB', 'Lower_BB', '%K', '%D', 'ATR']
print("Available columns in data:", data.columns)  # Debugging line

# Check if all required features are present
missing_features = [feature for feature in features if feature not in data.columns]
if missing_features:
    raise ValueError(f"Missing features: {missing_features}")

X = data[features]
y = data['Label']

# Convert categorical target to numerical
y = pd.get_dummies(y, drop_first=True)  # 1 for Buy, 0 for Sell (simplified)

# Scale the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Initialize TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)

# Initialize the Random Forest Classifier
model = RandomForestClassifier(n_estimators=500, max_depth=15, min_samples_split=10, min_samples_leaf=5, random_state=42)

# Perform cross-validation
for train_index, test_index in tqdm(tscv.split(X_scaled), total=tscv.n_splits, desc="Cross-validation"):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))

# Train the model on the entire dataset
model.fit(X_scaled, y)

# Generate buy/sell signals
data['Prediction'] = model.predict(X_scaled)

# Example plot
plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], label='Price', color='blue')
buy_signals = data[data['Prediction'] == 1]
sell_signals = data[data['Prediction'] == 0]
plt.scatter(buy_signals.index, buy_signals['Close'], marker='^', color='g', label='Buy Signal', alpha=1)
plt.scatter(sell_signals.index, sell_signals['Close'], marker='v', color='r', label='Sell Signal', alpha=1)
plt.title(f"{stock} Buy/Sell Signals")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='best')
plt.show()
```
